```python
def filt_rpkm_by_avg(rpkm, groups, cutoff_rpkm=4):
    import numpy as np
    import pandas as pd
    groups_unique = np.unique(groups)
    averages = []
    for i in groups_unique:
        averages.append(rpkm.loc[:, [val==i for val in groups]].mean(1))
    avg_df = pd.DataFrame(averages).T
    pass_filt = (avg_df > cutoff_rpkm).any(1)
    print('number of genes passing cutoff: %i' % pass_filt.sum())
    return(rpkm.loc[pass_filt, :])
```

### Differential gene analysis

```python
# import differential gene expression
diff_gene = pd.read_csv('./expression/diff_output.txt', sep='\t', index_col=0)
diff_gene.index.rename('RepeatID', inplace=True)
mm10_gene = diff_gene['Annotation/Divergence'].str.split('|').str[0]

# create gene name index
diff_gene['gene'] = mm10_gene
diff_gene = diff_gene.reset_index().set_index('gene').drop(labels='RepeatID', axis=1)

# filter entries without a gene
# diff_gene = diff_gene.loc[diff_gene.index.notna(), :]
print(diff_gene.shape)
```

```python
def pull_comparision(diff_gene, comp):
    return diff_gene.loc[:, [comp + ' Log2 Fold Change',
                            comp + ' p-value',
                            comp + ' adj. p-value']]

def count_p_fc(diff, pval=0.05, fc=2):
    
    tmp = [((diff.adj_pval <= pval) & (diff.log2fc >= np.log2(fc))).value_counts(),
            ((diff.adj_pval <= pval) & (diff.log2fc <= np.log2(fc))).value_counts()]
    print('up 2 vs. 1:')
    print(tmp[0])
    print('down 2 vs. 1:')
    print(tmp[1])
    return pd.DataFrame(tmp, index=['up', 'down']).fillna(0).T

# extract groups
comp_dict = {}
pattern='(\w* vs. \w*).*'
for col in diff_gene.columns.values:
    m = re.search(string=col, pattern=pattern)
    if m:
        df = pull_comparision(diff_gene, m.group(1))
        df.columns = ['log2fc', 'pval', 'adj_pval']
        comp_dict[re.sub('G0[0-9]_', '', m.group(1))] = df
```

Write gene lists and volcano plots for comparison

```python
#### Write gene list for analysis with metascape
pval = 0.05 # p value cutoff
log2fc = np.log2(2) # fc cutoff for interesting genes
fc_max = 1.3 # max fc to define as unchanged
up_down_dict = {}

if not os.path.isdir('./gene_lists'):
    os.mkdir('./gene_lists')

for key, diff in comp_dict.items():
    print('analyzing:', key)
    # count genes passing criteria
    count_p_fc(diff)
    
    # filter based on expression
    diff_filt = diff.loc[set(diff.index.tolist()).intersection(rpkm_filt.index.tolist()), :]
    
    up = ( diff_filt.adj_pval <= pval ) & \
    ( diff_filt.log2fc >= log2fc )
    print("sig up genes passing expression filter:", len(up[up]))
    
    dn = ( diff_filt.adj_pval <= pval ) & \
    ( diff_filt.log2fc <= -log2fc )
    print("sig dn genes passing expression filter:", len(dn[dn]))

    
    with open('./gene_lists/'+key+'_up_log2fc%.2f_p%.2f.txt' % (log2fc, pval), 'w') as f:
        f.write('gene\n')
        for val in up[up].index.values:
            f.write(val+'\n')
        f.close()

    with open('./gene_lists/'+key+'_dn_log2fc%.2f_p%.2f.txt' % (log2fc, pval), 'w') as f:
        f.write('gene\n')
        for val in dn[dn].index.values:
            f.write(val+'\n')
        f.close()
        
    up_down_dict[key[12:]] = ( len(dn[dn].index.values),
                               len(up[up].index.values) )
    
    # make volcano plot
    get_diff_volcano(diff, label_n_top_genes=10, plot_size=5)
    plt.title(key)
    plt.savefig('./figures/'+key+'_volcano.png', bbox_inches='tight')
    plt.close()
    print()
```
