```python
diff = pd.read_csv('./merge_peaks/diff_output.txt', sep='\t', index_col=0)
diff.index.rename('peakID', inplace=True)
```

```python
def pull_comparision(diff_gene, comp):
    return diff_gene.loc[:, [comp + ' Log2 Fold Change',
                            comp + ' p-value',
                            comp + ' adj. p-value']]

def count_p_fc(diff, pval=0.05, fc=2):
    
    tmp = [((diff.adj_pval <= pval) & (diff.log2fc >= np.log2(fc))).value_counts(),
            ((diff.adj_pval <= pval) & (diff.log2fc <= np.log2(fc))).value_counts()]
    print('up 2 vs. 1:')
    print(tmp[0])
    print('dn 2 vs. 1:')
    print(tmp[1])
    return pd.DataFrame(tmp, index=['up', 'down']).fillna(0).T

# extract groups
comp_dict = {}
pattern='(\w* vs. \w*).*'
for col in diff.columns.values:
    m = re.search(string=col, pattern=pattern)
    if m:
        df = pull_comparision(diff, m.group(1))
        df.columns = ['log2fc', 'pval', 'adj_pval']
        comp_dict[re.sub('G0[0-9]_', '', m.group(1))] = df
```

```python
for key, diff_comp in comp_dict.items():
    if 'creneg' in key:
        print(key)
        tmp = count_p_fc(diff_comp)
```
