```python
# import differential gene expression
diff_gene = pd.read_csv('./expression/diff_output.txt', sep='\t', index_col=0)
diff_gene.index.rename('RepeatID', inplace=True)
mm10_gene = diff_gene['Annotation/Divergence'].str.split('|').str[0]

# create gene name index
diff_gene['gene'] = mm10_gene
diff_gene = diff_gene.reset_index().set_index('gene').drop(labels='RepeatID', axis=1)

# filter entries without a gene
# diff_gene = diff_gene.loc[diff_gene.index.notna(), :]
print(diff_gene.shape)

# import differential gene expression
diff_gene_edger = pd.read_csv('./expression/diff_output_edger.txt', sep='\t', index_col=0)
diff_gene_edger.index.rename('RepeatID', inplace=True)
mm10_gene = diff_gene_edger['Annotation/Divergence'].str.split('|').str[0]
diff_gene_edger['gene'] = mm10_gene
diff_gene_edger = diff_gene_edger.reset_index().set_index('gene').drop(labels='RepeatID', axis=1)

# filter entries without a gene
# diff_gene = diff_gene.loc[diff_gene.index.notna(), :]
print(diff_gene_edger.shape)
```

```python
def pull_comparision(diff_gene, comp):
    return diff_gene.loc[:, [comp + ' Log2 Fold Change',
                            comp + ' p-value',
                            comp + ' adj. p-value']]

def count_p_fc(diff, pval=0.05, fc=2):
    
    tmp = [((diff.adj_pval <= pval) & (diff.log2fc >= np.log2(fc))).value_counts(),
            ((diff.adj_pval <= pval) & (diff.log2fc <= np.log2(fc))).value_counts()]
    print('up 1 vs. 2:')
    print(tmp[0])
    print('down 1 vs. 2:')
    print(tmp[1])
    return pd.DataFrame(tmp, index=['up', 'down']).fillna(0).T

# extract groups
comp_dict = {}
pattern='(\w* vs. \w*).*'
for col in diff_gene.columns.values:
    m = re.search(string=col, pattern=pattern)
    if m:
        df = pull_comparision(diff_gene, m.group(1))
        df.columns = ['log2fc', 'pval', 'adj_pval']
        comp_dict[re.sub('G0[0-9]_', '', m.group(1))] = df

comp_dict_edger = {}
pattern='(\w* vs. \w*).*'
for col in diff_gene_edger.columns.values:
    m = re.search(string=col, pattern=pattern)
    if m:
        df = pull_comparision(diff_gene_edger, m.group(1))
        df.columns = ['log2fc', 'pval', 'adj_pval']
        comp_dict_edger[re.sub('G0[0-9]_', '', m.group(1))] = df
```

```python
de = comp_dict_edger['control vs. dmhca'].sort_values('adj_pval')
```

```python
up = de.loc[(de.log2fc >= 1) & (de.adj_pval <= 0.05)].index.tolist()
with open('./expression/diff_expression/dmhca_up_fc1_fdr05.txt', 'w') as f:
    f.write('Gene\n')
    for gene in up:
        f.write(gene + '\n')
    f.close()
    
dn = de.loc[(de.log2fc <= 1) & (de.adj_pval <= 0.05)]
with open('./expression/diff_expression/dmhca_dn_fc1_fdr05.txt', 'w') as f:
    f.write('Gene\n')
    for gene in dn:
        f.write(gene + '\n')
    f.close()
```
