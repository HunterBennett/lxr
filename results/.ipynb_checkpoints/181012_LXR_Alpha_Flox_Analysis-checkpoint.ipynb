### Make volcano plots for each comparison

```python
#### Write gene list for analysis with metascape
pval = 0.05 # p value cutoff
log2fc = np.log2(2) # fc cutoff for interesting genes
fc_max = 1.3 # max fc to define as unchanged
up_down_dict = {}

if not os.path.isdir('./gene_lists'):
    os.mkdir('./gene_lists')

for key, diff in comp_dict.items():
    
    up = ( comp_dict[key].adj_pval <= pval ) & \
    ( comp_dict[key].log2fc >= log2fc )
    
    dn = ( comp_dict[key].adj_pval <= pval ) & \
    ( comp_dict[key].log2fc <= -log2fc )
    
    with open('./gene_lists/'+key+'_up_log2fc%.2f_p%.2f.txt' % (log2fc, pval), 'w') as f:
        f.write('gene\n')
        for val in up[up].index.values:
            f.write(val+'\n')
        f.close()

    with open('./gene_lists/'+key+'_dn_log2fc%.2f_p%.2f.txt' % (log2fc, pval), 'w') as f:
        f.write('gene\n')
        for val in dn[dn].index.values:
            f.write(val+'\n')
        f.close()
        
    up_down_dict[key[12:]] = ( len(dn[dn].index.values),
                               len(up[up].index.values) )
    
    # make volcano plot
    get_diff_volcano(diff, label_n_top_genes=10, plot_size=4)
    plt.title(key)
    plt.savefig('./figures/'+key+'_volcano.pdf', bbox_inches='tight')
    plt.close()
```

```python
def get_diff_volcano(df, sig_pval=0.05, pseudocount=10**-50, label_n_top_genes=10, fontsize=8):
    '''function takes output from differential gene analysis
    (df must contain log2fc column and adj_pval column)
    and returns a volcano plot. Accepts:
    df - data frame with log2fc and adj_pval (fdr corrected)
    sig_pval = pval cutoff for coloring points
    pseudocount = artificial ceiliing on pvalues to compress plot
    label_n_sig_genes = number of genes to label on scatter plot
    fontsize = size of labels text'''
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    
    df['log10p'] = -np.log10(df.adj_pval + pseudocount)
    df['sig'] = df.adj_pval <= sig_pval
    fig, ax = plt.subplots(figsize=(5,5))
    plt.scatter(
        x=df.log2fc,
        y=df.log10p,
        s=8,
        c=['r' if i else 'k' for i in df.sig],
        )
    
    # label axis
    ax.set_xlabel('log2fc')
    ax.set_ylabel('-log10p')
    
    # label top ranked genes
    for i, point in df.sort_values('pval').iloc[:label_n_top_genes].iterrows():
            # print(i, point)
            ax.text(point['log2fc']+.02,
                    point['log10p'],
                    str(i),
                    fontsize=fontsize)
    return fig, ax
```

```python
for key, df in comp_dict.items():
    fig, volc = get_diff_volcano(df,
                 pseudocount=10**-20,
                 label_n_top_genes=10,
                 fontsize=6)
    volc.set_title(key);
    fig.savefig('./figures/'+key+' volcano.pdf', bbox_inches='tight')
    plt.close()
```
